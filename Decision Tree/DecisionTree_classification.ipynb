```python
import numpy as nm  
import matplotlib.pyplot as plt
import pandas as pd 
import sklearn
```

```python
path = 'dataset\jeeves.csv'
```

```python
dataset= pd.read_csv(path)
```

```python
dataset
```

```python
dataset.info()
```

```python
outlook = {'Overcast' : 2, 'Sunny': 1, 'Rain': 0}
dataset["Outlook"] = dataset["Outlook"].map(outlook)

temp = {'Cool' : 2, 'Mild': 1, 'Hot': 0}
dataset["Temp"] = dataset["Temp"].map(temp)

humidity = {'High' : 2, 'Mild': 1, 'Normal': 0}
dataset["Humidity"] = dataset["Humidity"].map(humidity)

wind = {'Strong': 1, 'Weak': 0}
dataset["Wind"] = dataset["Wind"].map(wind)
```

```python
play = {'Yes' : 1, 'No' : 0}
dataset["Tennis?"] = dataset["Tennis?"].map(play)
```

```python
dataset
```

```python
x= dataset.iloc[:, 1:5].values  
y= dataset.iloc[:, 5].values 
```

```python
print(x,y)
```

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25, random_state=0)  
```

```python
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()
x_train= st_x.fit_transform(x_train)    
x_test= st_x.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier  
classifier= DecisionTreeClassifier(criterion='entropy', random_state=0)  
classifier.fit(x_train, y_train)
```

```python
DecisionTreeClassifier( class_weight=None, 
                        criterion='entropy', 
                        max_depth=None,
                        max_features=None, 
                        max_leaf_nodes=None,
                        min_impurity_decrease=0.0, 
                        # min_impurity_split=None,
                        min_samples_leaf=1, 
                        min_samples_split=2,
                        min_weight_fraction_leaf=0.0, 
                        # presort=False,
                        random_state=0, 
                        splitter='best' )
```

```python
y_pred= classifier.predict(x_test)
```

```python
print(y_pred)
```

```python
print(y_test)
```

```python
 
from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(y_test, y_pred)
```
